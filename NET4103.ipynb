```python
from abc import ABC
from abc import abstractmethod
import networkx as nx
import numpy as np
import progressbar
from sklearn.metrics.pairwise import cosine_similarity

class LinkPrediction(ABC):
    def __init__(self, graph):
        """
        Constructor

        Parameters
        ----------
        graph : Networkx graph
        """
        self.graph = graph
        self.N = len(graph)

    def neighbors(self, v):
        """
        Return the neighbors list of a node

        Parameters
        ----------
        v: int
            node id

        Return
        ------
        neighbors_list : python list
        """
        neighbors_list = self.graph.neighbors(v)
        return list(neighbors_list)

    @abstractmethod
    def fit(self):
        raise NotImplementedError("Fit must be implemented")


class CommonNeighbors(LinkPrediction):
    def __init__(self, graph):
        super(CommonNeighbors, self).__init__(graph)

    def fit(self):
        scores = []
        visited_edges=set()
        for u in self.graph.nodes():
          for v in self.graph.nodes():
            if u<v:
              visited_edges.add((u,v))
              u_neighbors = set(self.neighbors(u))
              v_neighbors = set(self.neighbors(v))
              score = len(u_neighbors.intersection(v_neighbors))
              scores.append(np.array([(u,v),score]))
        return np.array(scores)


class Jaccard(LinkPrediction):
    def __init__(self, graph):
        super(Jaccard, self).__init__(graph)

    def fit(self):
        scores = []
        visited_edges=set()
        for u in self.graph.nodes():
          for v in self.graph.nodes():
            if u<v:
              visited_edges.add((u,v))
              u_neighbors = set(self.neighbors(u))
              v_neighbors = set(self.neighbors(v))
              if len(u_neighbors) == 0 or len(v_neighbors) == 0:
                  score = 0
              else:
                  score = len(u_neighbors.intersection(v_neighbors)) / len(u_neighbors.union(v_neighbors))
              scores.append(np.array([(u,v),score]))
        return np.array(scores)


class AdamicAdar(LinkPrediction):
    def __init__(self, graph):
        super(AdamicAdar, self).__init__(graph)

    def fit(self):
        scores = []
        visited_edges=set()
        for u in self.graph.nodes():
          for v in self.graph.nodes():
            if u<v:
              visited_edges.add((u,v))
              u_neighbors = set(self.neighbors(u))
              v_neighbors = set(self.neighbors(v))
              common_neighbors = u_neighbors.intersection(v_neighbors)
              score = 0
              for z in common_neighbors:
                  z_neighbors = set(self.neighbors(z))
                  if len(z_neighbors) > 0:
                      score += 1 / np.log(len(z_neighbors))
              scores.append(np.array([(u,v),score]))
        return np.array(scores)
```

```python
def evaluate_method(G,arr,n):
  count = 0
  for elt in arr:
    if G.has_edge(elt[0][0],elt[0][1]):
      count+=1
  return count/n
```

```python
import random
import networkx as nx
def assess_methods(G):
  # assume the original graph is stored as Gfb
  f = 0.1  # fraction of edges to remove
  num_edges = G.number_of_edges()
  n = 100 #compute n as the number of removed edges
  edges_to_remove = random.sample(list(G.edges()), int(f*num_edges))

  # create a new graph with the selected edges removed
  Gfb_removed = nx.Graph()
  Gfb_removed.add_nodes_from(G.nodes())
  Gfb_removed.add_edges_from([e for e in G.edges() if e not in edges_to_remove])

  # get the results of the three algorithms
  results= {'CommonNeighbors':CommonNeighbors(Gfb_removed).fit(), 'Jaccard':Jaccard(Gfb_removed).fit(), 'AdamicAdar':AdamicAdar(Gfb_removed).fit()}
  results_sorted = {}

  # Sort each subarray of each key by the score
  for key, value in results.items():
    sorted_arr = value[np.argsort(value[:, 1])][::-1]
    sorted_arr = sorted_arr[:n]
    results_sorted[key] = sorted_arr 
    
  # calculate the confidence score of each method
  confidence = dict()
  for key, value in results_sorted.items():
    confidence[key]= evaluate_method(Gfb_removed, value, n)

  print(confidence)
  return confidence
```

## Question5 : Find missing labels with the label propagation algorithms


```python
import networkx as nx
import numpy as np

def label_propagation(G, labels, node_subset,max_iter=100):
    """
    Applies the label propagation algorithm to a given network G
    with initial node labels specified in the dictionary 'labels'.
    The algorithm runs for a maximum of 'max_iter' iterations.

    Args:
        G (networkx.Graph): A NetworkX graph object.
        labels (dict): A dictionary with node labels.
        max_iter (int): Maximum number of iterations.

    Returns:
        A dictionary of labels for each node in the graph.
    """

    # Create a copy of the original node labels
    node_labels = dict(labels)

    noeud_manquante=[]
    for i in node_labels:
      if node_labels[i] == -1:
        noeud_manquante.append(i)


    # Convert the graph to a matrix representation
    adj_matrix = nx.to_numpy_array(G)

    # Normalize the matrix rows
    D = np.diag(np.sum(adj_matrix, axis=1))
    D_inv = np.linalg.inv(D)
    P = np.matmul(D_inv, adj_matrix)

    # Apply label propagation
    for i in range(max_iter):
        for node in node_subset:  
          neighbors = list(G.neighbors(node))
          if len(neighbors) == 0:
              continue
          neighbor_labels = []
          for n in neighbors:
            if n in node_labels.keys():
                neighbor_labels.append(node_labels[n])
          if (len(neighbor_labels) != 0):
            a= np.bincount(neighbor_labels)
            node_labels[node] = np.argmax(a)
          else: 
            node_labels[node] = 0
    return node_labels
```

```python
from networkx.algorithms import group_degree_centrality
import random
import networkx as nx
import numpy as np


def Suppression_Attributes(G_Cal, percentage, attribute):
  num_nodes = len(G_Cal.nodes)
  node_subset = random.sample(list(G_Cal.nodes), int(num_nodes * percentage))
  # Remove node attributes from each of the subsets
  for node in node_subset:
      G_Cal.nodes[node].pop(attribute, None)
  return node_subset


```

```python
def accurrancy(full_labels_Dorm_10,recovered_labels , node_subset ):
  k=0 ;
  for i in node_subset : 
    if full_labels_Dorm_10[i] == recovered_labels[i]:
      k+=1
  return k/ len(node_subset)
```
